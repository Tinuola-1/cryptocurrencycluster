```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import hvplot.pandas
import plotly.figure_factory as ff
import plotly.express as px
```

```python
file_path =r"C:\Users\tinuo\OneDrive\Desktop\crypto_data.csv"
crypto_df = pd.read_csv(file_path, index_col =0)
crypto_df.tail(5)
```

```python
#Discard all cryptocurrencies that are not being traded
crypto_df = crypto_df[crypto_df.IsTrading]
crypto_df.tail(5)
```

```python
# Remove all cryptocurrencies that donâ€™t have a defined algorithm 
crypto_df = crypto_df.dropna(subset=["Algorithm"])
crypto_df.shape
```

```python
# Remove the IsTrading column.
crypto_df.drop("IsTrading", axis=1, inplace=True)
crypto_df.shape
```

```python
#Remove all rows that have at least one null value.
crypto_df = crypto_df.dropna()
crypto_df.shape
```

```python
#Filter for cryptocurrencies that have been mined
crypto_df.drop(crypto_df.loc[crypto_df['TotalCoinsMined'] <= 0].index, inplace=True)
crypto_df.shape
```

```python
# Store the names of all cryptocurrencies on a DataFramed named coins_name, then use the crypto_df.index as the index for this new DataFrame.
coins_name = pd.DataFrame(crypto_df, index=crypto_df.index, columns=['CoinName'])
coins_name
```

```python
#delete the `CoinName` from the original dataframe.
crypto_df = crypto_df.drop(['CoinName'], axis = 1)
crypto_df

#df=pd.read_csv("crypto_data.csv").drop(['CoinName'], axis = 1)
#df.head(5)
```

```python
X= pd.get_dummies(crypto_df)
X.head()
```

```python
X_scaled = StandardScaler().fit_transform(X)
print(X_scaled[0:5])
```

```python
#Perform dimensionality reduction with PCA
pca = PCA(n_components=3)
```

```python
#Transform PCA datainto a dataframe
X_pca = pca.fit_transform(X_scaled)
pcs_df = pd.DataFrame(data=X_pca, columns=["PC 1", "PC 2", "PC 3"], index=crypto_df.index)
pcs_df.head()
```

```python
#EXplained variance
pca.explained_variance_ratio_
```

```python
#Best value for K
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of K values
for i in k:
	km = KMeans(n_clusters=i, random_state=0)
	km.fit(pcs_df)
	inertia.append(km.inertia_)

# Create the elbow curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", xticks=k, title="Elbow Curve")
```

```python
#Predicting K clusters
model = KMeans(n_clusters=5, random_state=5)
model
model.fit(pcs_df)
predictions = model.predict(pcs_df)
pcs_df["class"] = model.labels_
pcs_df.head()
```

```python
#Can the cryptocurrencies be clustered together?
clustered_df = pd.merge(crypto_df, pcs_df, left_index=True, right_index = True)
clustered_df.head()
```

```python
#add the coin name back
clustered_df = pd.merge(clustered_df, coins_name, left_index=True, right_index = True)
clustered_df.head()
```
